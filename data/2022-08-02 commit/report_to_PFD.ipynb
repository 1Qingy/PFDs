#Useful columns

TDR_TDR_TDR: TOTAL deaths reported

Cases_Total: TOTAL Inquest concluded/adjourned

Inq_op_Total: TOTAL Inquests still open (Abroad and E&W)

PFD_reports: Number of Prevention of Future Deaths Reports issued

PM_Standard
Standard rate Post Mortems

PM_NonStandard
Non-standard rate Post Mortems

TDR_IQ_T
Total deaths reported with an inquest Male and female Total

TDR_PIQ_T
Total deaths reported where inquest is possible Total

PM_Hist
Post Mortems including Histology

PM_Tox
Post Mortems including Toxicology

PM_less_inv
Post Mortems involving less invasive techniques

PM_less_invasive_only
Post Mortems involving only less invasive techniques

PM_HO_forensic_pathologist
Post Mortems conducted by a HO forensic pathologist

PM_defence_lawyer_request
Post Mortems conducted following request by defence lawyer

Postmortem: investigation of the cause of death

There are 433 PFDs in Coroner juidiciary website. https://www.judiciary.uk/subject/prevention-of-future-deaths/

But in coroner statistics website (https://www.gov.uk/government/statistics/coroners-statistics-2019/coroners-statistics-2019-england-and-wales#inquests-opened), there are 440 in 2021. This means that in 2021, there are at least 7 PFD reports that are not uploaded.

```python
by_year['PFD/reports (%)'] = by_year['PFD_reports']/by_year['TDR_TDR_TDR']*100
by_year['PFD/Inquests (%)'] = by_year['PFD_reports']/by_year['TDR_IQ_T']*100
by_year['PFD/PM (%)'] = by_year['PFD_reports']/by_year['PM']*100
```

I have assumed "PM" = PM standard + PM non-standard. I have verifyied using this website that this is true: https://www.gov.uk/government/statistics/coroners-statistics-2019/coroners-statistics-2019-england-and-wales#inquests-opened
