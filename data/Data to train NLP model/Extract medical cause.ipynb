The below code goes into a PDF and checks if the coroner wrote the conclusion of the Inquest (“Accident/Misadventure; Alcohol / Drug Related; Industrial Disease; Lawful/Unlawful Killing; Natural Causes; Open; Road Traffic Collision; Still Birth; and Suicide”). If so, it displays it.

```python
pdf = '/content/G199-Cause-of-death-list.pdf'
# creating an object 
file = open(pdf, 'rb')

# creating a pdf reader object
fileReader = PyPDF2.PdfFileReader(file)
```

```python
pages = ''
for i in range(fileReader.numPages): 
  page = fileReader.pages[i]
  text = page.extract_text()
  text = text.replace('\n', '')
  pages = pages + " " + text
```
